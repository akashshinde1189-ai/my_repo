```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
from statsmodels.tsa.ar_model import AutoReg
data=[10,12,14,16,18,20,20,22]
```

```python
plt.plot(data)
```

```python
ar_model=AutoReg(data,lags=1)
ar_model_fit=ar_model.fit()
```

```python
#forcasting for 8, 9,and 10(three values)
forecast=ar_model_fit.predict(8,10)
print(forecast)
```

```python
forecast_list=forecast.tolist()#array converted to list
```

```python
my_data=data+forecast_list
```

```python
print(my_data)
```

```python
plt.plot(my_data)
```
